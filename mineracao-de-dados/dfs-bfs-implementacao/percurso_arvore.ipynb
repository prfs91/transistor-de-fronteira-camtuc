### 3. Execução dos Percursos Completos
Processamos a árvore a partir da raiz (Nó 'A') para identificar a ordem total de visitação de cada algoritmo. O resultado é exportado como imagem para garantir a integridade da formatação no relatório.

### 4. Busca Interativa por Nó Alvo
Nesta etapa, o usuário define um objetivo. O sistema calcula o caminho percorrido por cada algoritmo até encontrar o nó desejado, gerando um relatório detalhado e comparativos visuais.

```python
# Digitar o nó alvo da busca
no_alvo = input("Digite a letra do nó alvo para busca: ").upper()

# Processamento da busca específica
caminho_dfs, caminho_bfs = buscar_no_especifico(no_alvo)

if caminho_dfs:
    # 1. Gera a 'Ficha de Busca' em imagem
    salvar_busca_como_imagem(no_alvo, caminho_dfs, caminho_bfs)
    
    # 2. Exibe o comparativo gráfico lado a lado
    desenhar_comparativo_visual(caminho_dfs, caminho_bfs, no_alvo)
    
    # 3. Análise individual detalhada
    print(f"\n# DETALHAMENTO INDIVIDUAL: ALVO {no_alvo}\n")
    print(f"\nANÁLISE DA BUSCA EM PROFUNDIDADE (DFS) | ANÁLISE VERTICAL\n")
    desenhar_percurso_unico(caminho_dfs, no_alvo, "Análise Profundidade (DFS)")
    print(f"\nANÁLISE DA BUSCA EM LARGURA (BFS) | ANÁLISE HORIZONTAL\n")
    desenhar_percurso_unico(caminho_bfs, no_alvo, "Análise Largura (BFS)")
else:
    print(f"O nó '{no_alvo}' não foi encontrado na estrutura da árvore.")
```

### 5. Análise Técnica e Comparativa


| Característica | Busca em Profundidade (DFS) | Busca em Largura (BFS) |
| :--- | :--- | :--- |
| **Estratégia** | Exploração vertical (descida profunda). | Exploração horizontal (níveis/camadas). |
| **Estrutura Base** | Pilha (LIFO) ou Recursão. | Fila (FIFO). |
| **Vantagem** | Eficiente para encontrar folhas distantes. | Garante o caminho mais curto em níveis. |

**Conclusão Final:**
Como observado nos resultados visuais, a **DFS** tendeu a percorrer ramos inteiros antes de mudar de direção, enquanto a **BFS** mapeou a vizinhança imediata da raiz. Ambos operam em complexidade $O(V + E)$, sendo ferramentas fundamentais para a mineração de padrões em estruturas complexas.
